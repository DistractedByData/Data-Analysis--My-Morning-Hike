#### Libraries:

I am importing pandas to handle the CSV file containing the coordinates from my hike this morning.

Matplotlib and Plotly will both be used to visualize the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
data = pd.read_csv("./Morning hike.csv")
```

I plan to eventually plot the `Latitude` and `Longitude` visually to aid in interpreting them. However, initially, it could be interesting to analyze the elevation like the maximum, minimum, and total gain. 

```python
max_elevation = data.Elevation.max()
min_elevation = data.Elevation.min()
start_elevation = data.iloc[0].Elevation
```

```python
print(f"The total elevation gain during my hike was \
{max_elevation - min_elevation} meters. \n\
    - Highest elevation: {max_elevation} meters above sea level. \n\
    - Lowest Elevation: {min_elevation} meters above sea level.\n\
The start and end of the hike was at {start_elevation} meters above sea level.")
```

### Visualizing Elevation

```python
half_way = data.shape[0] / 2
first_quarter = data.shape[0] / 4
third_quarter = first_quarter * 3
thirdQ_elevation = data.iloc[int(third_quarter)].Elevation
firstQ_elevation = data.iloc[int(first_quarter)].Elevation
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(data.Elevation, color='orange', linestyle="-")
ax.axvline(half_way, color='blue', linestyle="--")
ax.axvline(first_quarter, color='red', linestyle="--")
ax.axvline(third_quarter, color='green', linestyle="--")

max_elevation = max(data.Elevation)

ax.text(half_way + 300, max_elevation - 20, 'Halfway\n57min', color='blue', ha='center')
ax.text(first_quarter - 360, max_elevation - 20, 'First Quarter\n28.5min', color='red', ha='center')
ax.text(third_quarter + 360, max_elevation - 20, 'Third Quarter\n85.5min', color='green', ha='center')

ax.text(1950, 1500, f'Elevation Change\nQ2 to Q3\n{max_elevation - thirdQ_elevation} meters')
ax.arrow(2200, 1490, dx=-280, dy=-100)

ax.text(-100, 1500, f'Elevation Change\nQ1 to Q2\n{max_elevation - firstQ_elevation} meters')
ax.arrow(200, 1490, dx=440, dy=-70)

ax.set_title("Elevation Gain (1690 meters Max Elevation)")
ax.set_xlabel("Coordinate Snapshots (1 every 3 seconds)")
ax.set_ylabel("Elevation (meters)")

plt.show()
```

It is interesting, but not surprising, to observe that more elevation is covered on the way down within the same time frame, specifically from quarter 2 (halfway) to quarter 3, compared to quarter 1 to quarter 2. Going downhill naturally tends to be quicker than climbing uphill.

## Visualize the Longitude and Latitude

### Scatterplot
Now that I have analyzed and visualized my hike in terms of elevation gain, I will proceed to map out the journey. This can be done from a bird's-eye view by plotting the longitude and latitude on a scatterplot. I will also include the elevation as a visual aid to display where the inclines and declines occur during the route.

```python
# Extract latitude, longitude, and elevation columns
latitude = data['Latitude']
longitude = data['Longitude']
elevation = data['Elevation']

fig, ax = plt.subplots(figsize=(8,6))

# Create a scatter plot
scatter = ax.scatter(longitude, latitude, c=elevation, cmap='viridis')
plt.colorbar(scatter, label='Elevation (m)')

# Set plot title and labels
ax.set_title('Morning Hike: Latitude, Longitude, and Elevation')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display the plot
plt.show()
```

It's fascinating to see the longitude and latitude plotted visually! It becomes apparent that the distance covered when the elevation is consistent is much greater than the ground covered during elevation changes, highlighting how elevation can significantly impact the overall distance traveled during a hike.

### 3D Scatterplot

I think it would be interesting to compare the 2D scatterplot with a 3D version, considering the fact that the dimensions being plotted are spatial in nature. Visually plotting them with a 3D scatterplot makes sense to me.

```python
# Create an interactive 3D scatter plot
fig = px.scatter_3d(data, x='Longitude', y='Latitude', z='Elevation', color='Elevation')

# Set plot title and labels
fig.update_layout(title='Morning Hike: Latitude, Longitude, and Elevation (3D Scatter Plot)',
                  scene=dict(xaxis_title='Longitude', yaxis_title='Latitude', zaxis_title='Elevation'))

# Display the plot
fig.show()


# Save as an interactive HTML file
fig.write_image("scatter_map.png")
```

### Google Maps Views

Finally, I think it would be nice to see the entire hike plotted on top of a map. This will help put everything into perspective.

```python
# Calculate the center coordinates
center_lat = data['Latitude'].mean()
center_lon = data['Longitude'].mean()

# Create an interactive 3D scatter plot on Google Earth map with default view
fig = px.scatter_mapbox(data, lat='Latitude', lon='Longitude', 
                        height=600, zoom=1, hover_name='Elevation')

# Set map style to Google Earth
fig.update_layout(mapbox_style='open-street-map')

# Set the center coordinates and zoom level
fig.update_layout(mapbox_center={'lat': center_lat, 'lon': center_lon})
fig.update_layout(mapbox_zoom=13.5)

# Set plot title
fig.update_layout(title='Morning Hike: Tunnel Mountain in Banff National Park')

# Display the plot
fig.show()

# Save as an interactive HTML file
pio.write_html(fig, "scatter_map.html")
```
